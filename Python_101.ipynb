```python
# The Currency Exchange
inr = 75000
yen = inr * 1.75
print("Rs " + str(inr) + " converts to Yen " + format(yen,".2f"))
```

```python
# The Grocery Bill
bread = 40
milk = 65.50
eggs = 120
total = bread + milk + eggs
avg = total / 3
print("Total: Rs " + str(format(total,".2f")))
print("Average per item: Rs" + str(format(avg,".2f")))
```

```python
# The Password Strength Check
password = "hello@2026"
length_ok = len(password) >= 8
contains_at = "@" in password
strong = length_ok and contains_at

print("Password:",password)
print("Length OK(>=8):",length_ok)
print("Contains '@':",contains_at)
print("Strong:",strong)
```
